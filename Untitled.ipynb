```python
from matplotlib import pyplot as plt
import numpy as np
import keras
from sklearn.metrics import roc_auc_score, roc_curve, auc
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.regularizers import l2
import pandas as pd
from keras.layers.normalization import BatchNormalization
from sklearn import datasets
```

```python
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()
y=data.target
```

```python
X=data.data
```

```python
X.shape
```

```python
from keras import regularizers, optimizers
from keras.constraints import max_norm
from keras import initializers, optimizers

model2 = Sequential()
model2.add(Dense(10, activation='relu', input_shape=(30,)))

model2.add(Dense(30, activation='relu'))
model2.add(Dense(30, activation='relu'))


model2.add(Dense(1, activation='sigmoid'))

sgd = optimizers.SGD(lr=0.01, decay=0.00001, momentum=0.9, nesterov=True)
model2.compile(loss='binary_crossentropy', optimizer=sgd, metrics=['accuracy'])

import os
import time
from keras.callbacks import TensorBoard

log_dir = './tb_log/' + time.strftime("%c")
log_dir = log_dir.replace(' ', '_').replace(':', '-')
if not os.path.exists(log_dir):
    os.makedirs(log_dir)
tb = TensorBoard(log_dir=log_dir,
                 histogram_freq=1,
                 write_graph=True,
                 write_grads=False,
                 write_images=False)


callbacks=keras.callbacks.EarlyStopping(monitor='val_loss',min_delta=0.00001, patience=10, mode='auto')


history2 = model2.fit(X, y, batch_size=1, epochs=500, verbose=1,
                shuffle = True, callbacks=[callbacks])

```
